Este es un data Set, que trata sobre el futbol argentino, sobre los torneos de primera division que se disputaron entre el años 2015 al 2022 incluidos.
Este trabajo tiene como objetivo hacer predecciones de resultados, teniendo en cuenta variables como posesion de pelota, remates al arco, tarjetas, posesion de pelota y demas variables.

DESCRIPCION VARIABLES



'torneo', --> Nombre del Torneo es Disputa.
'fecha', --> Numero de la Fecha, Ej: Fecha 1.
'partido',--> Numero de partido de la fecha.
'equipo_local',--> Nombre del equipo que juega de local.
'equipo_visitante',--> Nombre del equipo que juega de visitante.
'goles_local', --> cantidad de goles del equipo local.
'goles_visitante', --> cantidad de goles del equipo visitante.
'posesion_local',--> cantidad de porcentaje te tenencia del balon del equipo local.
'tiros_arco_local',--> disparos al arco del equipo local.
'intentos_local', --> disparos del equipo local ( no necesariamente al arco en situacion de gol).
'faltas_local', --> cantidad de faltas o infracciones del equipo local.
'tiro_esquina_local',--> lanzamientos de esquina del equipo local.
'posesion_visitante', --> cantidad de porcentaje te tenencia del balon del equipo visitante.
'tiros_arco_visitante',--> disparos al arco del equipo visitante.
'intentos_visitante',-->  disparos del equipo visitante ( no necesariamente al arco en situacion de gol).
'faltas_visitante',--> infracciones hechas por equipo visitante 
'tiro_esquina_visitante',--> lanzamientos de esquina del equipo visitante.
'amarillas_local',--> amarillas del equipo local. 
'amarillas_visitante',--> amarillas del equipo visitante.
'rojas_local',--> tarjetas rojas (expulsion) del equipo local.
'rojas_visitante',--> tarjetas rojas (expulsion) del equipo visitante.
'valor_mercado_local',--> El valor de mercado del equipo local según el sitio web Transfermarkt.
'altura_media_local',--> Promedio de altura del equipo local.
'edad_media_local',--> Promedio de edad del equipo local.
'proporcion_zurdos_local',--> Proporcion de jugadores que tienen la zurda como pierna habil del equipo local.
'valor_mercado_visitante',--> El valor de mercado del equipo visitante según el sitio web Transfermarkt.
'altura_media_visitante',--> Promedio de altura del equipo visitante.
'edad_media_visitante',--> Promedio de edad del equipo  visitante.
'proporcion_zurdos_visitante',--> Proporcion de jugadores que tienen la zurda como pierna habil del equipo visitante.
'resultado',--> Resultado de como salio el encuentro.
'fecha_encuentro',--> Fecha del partido entre los equipos.
'apuesta_local',--> Cuota que paga la victoria del equipo local.
'apuesta_visitante',--> Cuota que paga la victoria del equipo  visitante.
'apuesta_empate'--> Cuota que paga el empate.

Gracias a las variables proporcionadas, aplicando ciencias de datos podemos predecir, entre otros,  el resultado de los partidos de fútbol argentino, lo que beneficia a una variedad de partes interesadas.

Las variables incluyen información sobre el torneo, la fecha, el partido, los equipos, los goles, la posesión, los tiros, los intentos, las faltas, los tiros de esquina, las tarjetas amarillas, las tarjetas rojas, el valor de mercado, la altura, la edad, la pierna hábil y el resultado.
