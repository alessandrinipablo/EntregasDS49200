Este conjunto de datos se de la pagina kaggle.com, el cual trate todas las caracteristicas y actuaciones de los jugadores en sus respectivas ligas.

Algunas de las  variables comprenden 'valor_más alto (precio de la ficha)', 'juegos disputados', 'minutos jugados','asistencias (pases goles)', 'goles' entre otros.

Con las variables que presenta el df, tal lo nombrado y tambien descripto en las columnas, podemos llegar a conclusion de:

*Predecir el rendimiento futuro de los jugadores.
*Identificar jugadores con potencial.
*Comparar jugadores.
*Analizar el rendimiento de los equipos.
*Analisis fisicos de los jugadores, para prevenir lesiones, y sus implicancias.

Estas acciones tambien ayudan y beneficia a clubes de futbol, para saber que jugadores comprar, dinero a invertir, proyecto deportivos de largo plazo, invertir en ideas de modelos de futbol en divisiones inferiores.






DESCRIPCION DE VARIABLES DEL DATASET 

-player: jugador
-team: equipo
-name: nombre
-position: posición (en el campo)
-height: altura
-age: edad
-appearance: apariciones (juegos disputados)
-goals: goles
-assists: asistencias
-yellow cards: tarjetas amarillas
-second yellow cards: segundas tarjetas amarillas (que llevan a roja)
-red cards: tarjetas rojas
-goals conceded: goles recibidos
-clean sheets: partidos con la valla invicta
-minutes played: minutos jugados
-days_injured: días de baja por lesión
-games_injured: partidos perdidos por lesión
-award: premio
-current_value: valor actual (de mercado)
-highest_value: valor máximo alcanzado
-position_encoded: posición codificada, ejemplo juega de posicion x, con x de 1 a 11( sirve para análisis estadístico)
-winger: extremo (jugador que juega sobre las bandas)
